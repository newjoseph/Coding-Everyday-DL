```
#import library

import pandas as pd
import tensorflow as tf

#import data from a csv file
path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
boston = pd.read_csv(path)

print(boston.head())
```

```
#separate independent and dependent variables

boston_ind = boston[['crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptratio','b','lstat']]
boston_de = boston[['medv']]

print(boston_ind.shape , boston_de.shape)
```

```
# make a model structure
X = tf.keras.layers.Input(shape=[13])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse') #mse is mean squared error

#fit the model with the given data
model.fit(boston_ind, boston_de, epochs=1000)
```

```
# Utilize the model
print(boston_de)

#compare the first five
model.predict(boston_ind[0:5])
boston_de[0:5]

#check the model's fomula
model.get_weights()
```
