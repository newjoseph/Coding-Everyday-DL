```
# categorical data 
# classification 


#import library
import pandas as pd
import tensorflow as tf


#import data from a csv file.
path = "https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv"

iris = pd.read_csv(path)
iris.head()
```

```
#one-hot encoding

encoding = pd.get_dummies(iris)
encoding.head()
```

```
# separate independent and dependent variable

print(encoding.columns)

ind =encoding[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
de = encoding[['품종_setosa', '품종_versicolor',
       '품종_virginica']]


print(ind.shape, de.shape)
```

```
#make a model structure

X = tf.keras.layers.Input(shape=[4])
Y = tf.keras.layers.Dense(3, activation= 'softmax')(X)
model = tf.keras.models.Model(X,Y)

model.compile(loss= 'categorical_crossentropy', metrics='accuracy')

#fit the model with the given data
model.fit(ind, de, epochs= 100)
```

```
#utilize the model

model.predict(ind[0:5])
```
