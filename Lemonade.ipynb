```
#import library
import pandas as pd
import tensorflow as tf

#import data from a csv file
path =  'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'

lemon = pd.read_csv(path)

#check data shape
print(lemon.shape)

# separate independent and dependent variables

lemon_ind = lemon[['온도']]
lemon_de = lemon[['판매량']]

print(lemon_ind.shape, lemon_de.shape)

# make a model structure
X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')

print(X)
```

```
# fit the model with the given data

model.fit(lemon_ind, lemon_de, epochs = 1000)
```

```
# Utilize the model

print(lemon_de)
model.predict(lemon_ind)
```
