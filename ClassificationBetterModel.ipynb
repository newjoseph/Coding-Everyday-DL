```
# Better learning model for classification using Iris data

#import library
import pandas as pd
import tensorflow as tf

#import data from a csv file.
path = "https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv"
iris = pd.read_csv(path)
iris.head()
```

```
# one-hot encoding

encoding = pd.get_dummies(iris)
encoding.head()

#separate the variables

ind = encoding[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
de = encoding[['품종_setosa', '품종_versicolor', '품종_virginica']]
print(ind.shape, de.shape)
```

```
# make a model structure using BatchNomalization

# this is the previous version
# X = tf.keras.layers.Input(shape=[4])
# H = tf.keras.layers.Dense(8, activation='swish')(X)
# H = tf.keras.layers.Dense(8, activation='swish')(H)
# H = tf.keras.layers.Dense(8, activation='swish')(H)
# Y = tf.keras.layers.Dense(3, activation='softmax')(H)


X = tf.keras.layers.Input(shape=[4])


H = tf.keras.layers.Dense(8)(X)  # just make 8 perceptrons (as I choose)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(3, activation ='softmax')(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss = 'categorical_crossentropy', metrics='accuracy')
```

```
# fit the model

model.fit(ind, de, epochs=1000)
```
